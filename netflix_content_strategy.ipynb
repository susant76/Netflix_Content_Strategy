```python
# Q. Which genres are most common in Movies and TV Shows?
```

```python
genre_type = (
    df.assign(listed_in=df['listed_in'].str.split(','))
      .explode('listed_in')
)

genre_type['listed_in'] = genre_type['listed_in'].str.strip()

top10 = genre_type['listed_in'].value_counts().head(10).index

genre_comparison = (
    genre_type[genre_type['listed_in'].isin(top10)]
    .groupby(['listed_in', 'type'])
    .size()
    .unstack(fill_value=0)
)

genre_comparison
```

```python
genre_comparison.plot(
    kind='bar',
    figsize=(12,6)
)                                                               ## visuals

plt.title("Movies vs TV Shows Across Top Genres")
plt.xlabel("Genre")
plt.ylabel("Number of Titles")

plt.xticks(rotation=45)

plt.show()
```

# Business Insights & Strategic Recommendations


## Summary

The objective of this project was to analyze Netflix's global content library and identify patterns that could support strategic business decisions.

After cleaning the data, performing feature engineering, and conducting exploratory data analysis, several key insights were identified regarding Netflix's content strategy, audience preferences, market presence, and content acquisition approach.




## 1. Content Strategy

### Observation
- Movies constitute the majority of Netflix's content library.
- TV Shows represent a smaller but significant portion of the catalog.

### Business Interpretation
- Netflix relies heavily on movies because they generally require lower long-term viewer commitment and can be produced or licensed more quickly than multi-season TV Shows.

### Recommendation
- Continue investing in movies while selectively expanding high-quality TV Shows that improve long-term subscriber retention.




## 2. Content Growth

### Observation
- Content additions increased rapidly until the late 2010s before slowing in recent years.

### Business Interpretation
- Netflix appears to have shifted its focus from rapidly expanding its catalog to investing in higher-quality original productions.

### Recommendation
- Prioritize quality and audience engagement instead of increasing the number of titles alone.




## 3. Global Content Distribution

### Observation
- The United States and India contribute the largest share of Netflix's content library.

### Business Interpretation
- These countries represent Netflix's strongest production ecosystems and key subscriber markets.

### Recommendation
- Continue strengthening partnerships in major markets while increasing investments in emerging regions to diversify the content catalog.




## 4. Audience Strategy

### Observation
- Most titles are rated TV-MA and TV-14.

### Business Interpretation
- Netflix primarily targets adult and teenage audiences.

### Recommendation
- Continue investing in mature content while gradually expanding family-friendly and children's programming.




## 5. Genre Strategy

### Observation
- Dramas, International Movies, and Comedies are the most common genres.

### Business Interpretation
- These genres appeal to a broad global audience and contribute significantly to viewer engagement.

### Recommendation
- Maintain investment in successful genres while experimenting with underrepresented genres to attract niche audiences.




## 6. Content Acquisition Strategy

### Observation
- Many titles are added to Netflix several years after their original release.

### Business Interpretation
- Netflix continues to rely on both licensed and original content.

### Recommendation
- Maintain a healthy balance between licensed content and original productions to reduce long-term licensing dependency.




## 7. Creator Partnerships

### Observation
- Several directors and actors appear repeatedly across Netflix titles.

### Business Interpretation
- Netflix has established ongoing collaborations with successful creators.

### Recommendation
- Strengthen relationships with high-performing creators while identifying emerging talent for future productions.

## Final Takeaways

Based on this analysis, the following trends stand out:

• Movies dominate Netflix's content library.

• Content additions grew rapidly before stabilizing in recent years.

• The United States and India are the largest contributors.

• Dramas, Comedies, and International Movies are the most common genres.

• Netflix primarily targets mature audiences.

• Most TV Shows have only one season.

• A relatively small group of directors and actors contribute multiple titles.

• Netflix relies on both licensed and original content to build its catalog.
